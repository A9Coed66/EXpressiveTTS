```python
import os
```

```python
import librosa

folder_path = "/home/tuannd/tuanlha/Dataset/HaveASip"
duration_list = []
for episode in os.listdir(folder_path):
    audio_path = os.path.join(folder_path, episode)
    duration = librosa.get_duration(filename=audio_path)
    duration_list.append(duration)
len(duration_list)
```

```python
sum(duration_list)
```

```python
import matplotlib.pyplot as plt

# Tạo biểu đồ histogram cho duration_list
plt.figure(figsize=(10, 6))
plt.hist(duration_list, bins=100, edgecolor='black')
plt.title('Distribution of Audio Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
