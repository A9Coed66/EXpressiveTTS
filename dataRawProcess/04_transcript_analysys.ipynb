```python
import json
import os
def read_script(filename):
    """Read a script file and return its content."""
    with open(filename, 'r') as file:
        return file.read()
```

## Kiểm tra số lượng tệp denoise và transcript

```python
import os
denoise_path = '/home4/tuanlha/EXpressiveTTS/dataRawProcess/06_transcript'
transcript_path = '/home4/tuanlha/EXpressiveTTS/dataRawProcess/04_denoise'
playlist_name = 'CDMediaWhy'

for episode in os.listdir(os.path.join(transcript_path, playlist_name)):
    episode_script_path = os.path.join(transcript_path, playlist_name, episode)
    #calculte num file
    num_files_script = len([f for f in os.listdir(episode_script_path) if f.endswith('.lab')])
    episode_denoise_path = os.path.join(denoise_path, playlist_name, episode)
    num_files_denoise = len([f for f in os.listdir(episode_denoise_path) if f.endswith('.wav')])
    if num_files_denoise!= num_files_script:
        print(f"Mismatch in {playlist_name}/{episode}: {num_files_denoise} denoise files vs {num_files_script} script files")
```

```python
import os
import random
import csv

def pick_random_file(directory_path, num=5):
    file_path = [os.path.join(directory_path, file) for file in os.listdir(directory_path) if file.endswith('.wav')]
    if num>len(file_path):
        return file_path
    picked_files = random.sample(file_path, num)
    return picked_files

def create_random_file_list(root, num=5):
    channels = [dir for dir in os.listdir(root)]
    file_list = []

    for channel in channels:
        channel_path = os.path.join(root, channel)
        for episode in os.listdir(channel_path):
            episode_path = os.path.join(channel_path, episode)
            files = pick_random_file(episode_path, num)
            file_list.extend(files)

    return file_list
```

```python
data_list = create_random_file_list("/home4/tuanlha/EXpressiveTTS/dataRawProcess/05_transcript")
dct = {}
cnt = 0
for file in data_list:
    txt_path = file.replace('.wav', '.lab')
    with open(txt_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    dct[cnt] = {'file_path': file, 'mad1999/pho-whisper-large-ct2': lines[0]}
    cnt += 1
import json
with open('file_list.json', 'w', encoding='utf-8') as f:
    json.dump(dct, f, indent=4, ensure_ascii=False)
```

```python
import torch, torchaudio
from ichigo_asr.demo.utils import load_model

# Load Ichigo Whisper
ichigo_model = load_model(
        ref="homebrewltd/ichigo-whisper:merge-medium-vi-2d-2560c-dim64.pth",
        size="merge-medium-vi-2d-2560c-dim64",
)
device = "cuda:2" if torch.cuda.is_available() else "cpu"
ichigo_model.ensure_whisper(device)
ichigo_model.to(device)
```

```python
from vinorm import TTSnorm
S=TTSnorm("Hàm này được phát triển từ 8/2019. Có phải tháng 12/2020 đã có vaccine phòng ngừa Covid-19 xmz ?")
S
```

```python
import json
with open('./file_list.json', 'r', encoding='utf-8') as f:
    dct = json.load(f)
cnt = 0
for key, value in dct.items():
    wav, sr = torchaudio.load(value['file_path'])
    if sr != 16000:
       wav = torchaudio.functional.resample(wav, sr, 16000)
    transcribe = ichigo_model.inference(wav.to(device))
    text = transcribe[0].text
    dct[key]['ichigo-whisper'] = TTSnorm(text, lower = True).replace(' .', '').replace(' ,','')
    print(cnt)
    cnt+=1
with open('file_list.json', 'w', encoding='utf-8') as f:
    json.dump(dct, f, indent=4, ensure_ascii=False)
#41s
```
